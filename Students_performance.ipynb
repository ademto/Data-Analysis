```python
# Import Libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
# Load the data
data = pd.read_csv('StudentsPerformance.csv')

# Display the first few rows of the dataset
data.head()
```

### Explore the data

```python
# Display basic information about the dataset
data.info()
```

```python
data.describe(include='all')
```

### Data Cleaning

```python
# Check for missing values
data.isnull().sum()
```

```python
# Check for duplicates values
data.duplicated().sum()
```

## Data Analysis

```python
# Gender distribution
gender_distribution = data['gender'].value_counts()
```

### Categorical Variables Analysis

```python
plt.figure(figsize=(15, 10))

plt.subplot(2, 3, 1)
sns.countplot(x='gender', data=data)
plt.title('Gender Distribution')

plt.subplot(2, 3, 2)
sns.countplot(x='race/ethnicity', data=data)
plt.title('Race/Ethnicity Distribution')

plt.subplot(2, 3, 3)
sns.countplot(x='parental level of education', data=data, palette='viridis')
plt.title('Parental Education Distribution')
plt.xticks(rotation=45, ha='right')

plt.subplot(2, 3, 4)
sns.countplot(x='lunch', data=data)
plt.title('Lunch Distribution')

plt.subplot(2, 3, 5)
sns.countplot(x='test preparation course', data=data)
plt.title('Test Preparation Course Distribution')

plt.tight_layout()
plt.show()
```

### Distribution of Scores

```python
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.hist(data['math score'], bins=20, edgecolor='black', alpha=0.7)
plt.title('Distribution of Math Scores')

plt.subplot(1, 3, 2)
plt.hist(data['reading score'], bins=20, edgecolor='black', alpha=0.7)
plt.title('Distribution of Reading Scores')

plt.subplot(1, 3, 3)
plt.hist(data['writing score'], bins=20, edgecolor='black', alpha=0.7)
plt.title('Distribution of Writing Scores')

plt.tight_layout()
plt.show()
```

### Correlation Matrix

```python
# Explore the correlation between numeric variables.
correlation_matrix = data[['math score', 'reading score', 'writing score']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation between Scores')
plt.show()
```

### Pairplot

```python
# Visualize relationships between pairs of numeric variables.
sns.pairplot(data[['math score', 'reading score', 'writing score']])
plt.suptitle('Pairplot of Scores', y=1.02)
plt.show()
```
